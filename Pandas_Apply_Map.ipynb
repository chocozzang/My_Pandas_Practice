```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv")
```

```python
#df.head(10)
```

```python
Ans = df.shape

# 데이터 프레임의 행과 열 개수를 확인하려면 .shape를 활용하면 된다.
```

```python
Ans
```

```python
mapping = { "Unknown" : 'N',
           "Less than $40K" : 'a',
           "$40K - $60K" : 'b',
           "$60K - $80K" : 'c',
           "$80K - $120K" : 'd',
           "$120K +" : 'e'}

Ans2 = df
Ans2['newIncome'] = Ans2['Income_Category'].map(lambda x : mapping[x])
Ans2 = Ans2['newIncome']
```

```python
Ans2.head(4)
```

```python
mapping2 = { "Unknown" : 'N',
           "Less than $40K" : 'a',
           "$40K - $60K" : 'b',
           "$60K - $80K" : 'c',
           "$80K - $120K" : 'd',
           "$120K +" : 'e'}

Ans3 = df
Ans3['newIncome'] = Ans3['Income_Category'].apply(lambda x : mapping2[x])
Ans3 = Ans3['newIncome']
```

```python
Ans3.head(4)
```

```python
Ans4 = df

Ans4['AgeState'] = Ans4['Customer_Age'].apply(lambda x : x // 10 * 10)

# map이나 apply에는 (lambda x : x에 대한 식) 으로 나타낼 수 있다.
```

```python
Ans4.groupby('AgeState').size()
```

```python
Ans5 = df

def gradu(x):
  if 'Graduate' in x:
    return 1
  else:
    return 0

Ans5['newEduLevel'] = Ans5['Education_Level'].apply(gradu)

# 또한, apply에는 함수명을 매개변수로 사용할 수도 있는데,
# 이 함수는 구문 이전에 미리 정의해 두어야 하며 여러 조건을 지정할 수도 있다.
```

```python
Ans5.groupby('newEduLevel').size()
```

```python
Ans6 = df

def cre_lim(x):
  if x >= 4500: return 1
  else:         return 0


Ans6['newLimit'] = Ans6['Credit_Limit'].apply(cre_lim)
```

```python
Ans6.newLimit.value_counts()

# Ans6.groupyby('newLimit').size()로 출력하면 0/1 순서대로 출력한다.
# 해당 칼럼에 접근하여 .value_counts()로도 빈도를 카운팅할 수도 있다.
```

```python
Ans7 = df

def state_func(x):
  if x.Marital_Status == "Married" and x.Card_Category == "Platinum":
      return 1
  
  else:
    return 0

Ans7['newState'] = Ans7.apply(state_func, axis = 1)

Ans7 = Ans7.newState.value_counts() 

# 여러 개의 컬럼에 적용하고자 할 때는 df.apply(func, axis = 0 or 1) 로 대입할 수 있다.
# 이때 axis = 0은 행이고 axis = 1은 컬럼에 대해서 apply하는 것이다.
```

```python
Ans7
```

```python
Ans8 = df

def checkgender(gen):
  if gen == "M": return "male"
  else         : return "female"

Ans8.Gender = Ans8.Gender.apply(checkgender)

Ans8 = Ans8.Gender.value_counts()
```
