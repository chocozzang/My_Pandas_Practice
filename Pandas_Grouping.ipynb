```python
import pandas as pd
```

```python
DataURL = "https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv"
df = pd.read_csv(DataURL)
```

```python
df.head(5)
```

```python
Ans = df.host_name.value_counts().sort_index()

# 빈도수에 대한 이름별 정렬을 원하면, sort_index()로 정렬할 수 있다.
```

```python
Ans.head(5)
```

```python
Ans2 = df.host_name.value_counts().sort_values(ascending = False)

Ans2next = pd.DataFrame({Ans2.name : Ans2.index, 'counts' : Ans2.values})

Ans2next = Ans2next.set_index('host_name')
```

```python
Ans2next.head(5)
```

```python
Ans2good = df.groupby('host_name').size().to_frame().rename(columns = {0 : 'counts'}).sort_values('counts', ascending = False)

# df.gruopby : 데이터프레임을 host_name 기준으로 묶는다.
# .size() : 묶은 host_name의 크기를 각각 구한다.
# .to_frame() : Series 형태를 DataFrame으로 변경하는데, columns가 0으로 고정되어 있다.
# .rename(columns = {0 : 'counts'}) : 칼럼의 이름을 0에서 counts로 변경한다.
# .sort_values('counts', ascending = False) : counts를 기준으로 값을 정렬하고, 내림차순 이므로 ascending 조건은 False로 둔다.
```

```python
Ans2good.head(5)
```

```python
Ans3 = df.groupby(['neighbourhood_group', 'neighbourhood'], as_index = False).size()

# groupby는 여러 개의 조건을 걸 수 있으며 [] 안에 ,로 연결하면 된다.
# 또한 새로운 인덱스를 적용하고자 한다면 as_index = False면 된다.
# 아니라면 제일 앞에 있는 칼럼의 키가 인덱스가 된다.
```

```python
Ans3.head(4)
```

```python
Ans4 = Ans3.groupby('neighbourhood_group', as_index = False).max()
Ans5 = (df.groupby(['neighbourhood_group', 'neighbourhood'], as_index = False).size()).groupby('neighbourhood_group', as_index = False).max()
```

```python
Ans8 = df.groupby(['neighbourhood', 'neighbourhood_group'])['price'].mean()
```

```python
Ans8
```

```python
Ans9 = df.groupby(['neighbourhood', 'neighbourhood_group']).price.mean().unstack()

# .unstack()은 마지막 인덱스 즉, 제일 뒤에 grouping된 인덱스를 칼럼으로 올려주는 역할을 가지고 있다.
```

```python
Ans9
```

```python
Ans10 = Ans9

Ans10[Ans10.isnull()] = -999
Ans10_good = Ans9.fillna(-999)

# 직접 접근하여 값을 바꿀 수도 있겠지만,
# .fillna(x)를 활용하면 곧바로 NaN을 x로 변경할 수 있을 것이다.
```

```python
Ans12 = df[['room_type', 'neighbourhood_group']].groupby(['neighbourhood_group', 'room_type'], as_index = True).size().unstack()
Ans12.loc[:, :] = Ans12.values / Ans12.sum(axis = 1).values.reshape(-1, 1)

# sum에서 axis가 0이면 행을, 1이면 열을 더한다.
# reshape는 리스트의 데이터 구조를 변경하는 것이다.

# 이렇게 계산한 값을 loc[:, :] 즉, 데이터프레임 전체에 할당할 수 있다.
```
