```python
import pandas as pd
```

```python
dataURL = 'https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv'

df = pd.read_csv(dataURL, index_col = 0)
```

```python
df1 = df.iloc[:4, :]
df2 = df.iloc[4:, :]

display(df1)
display(df2)
```

```python
Ans = pd.concat([df1, df2], axis = 0)

# 데이터프레임의 병합 : pd.concat([병합할 프레임], axis = 0/1)
# axis가 0이면 위아래로 병합, 1이면 좌우로 병합한다.
```

```python
Ans
```

```python
df3 = df.iloc[:2, :4]
df4 = df.iloc[5:, 3:]

display(df3)
display(df4)
```

```python
Ans2 = pd.concat([df3, df4], axis = 0, join = 'inner')

# concat에서 공통 열 만을 병합하고자 한다면 join 옵션에 'inner'을 두면 되고,
# 전체를 하고자 하면 'outer' 옵션을 두면 된다.
```

```python
Ans2
```

```python
Ans3 = pd.concat([df3, df4], axis = 0).fillna(0)
```

```python
Ans3
```

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

```python
Ans4 = pd.merge(df5, df6, on='Algeria', how = 'inner')

# merge를 활용한 병합
# pd.merge(left, right, on = '공통된 이름이 있는 경우', how = '조인 방법')
# pd.merge(left, right, left_on = '좌측 열', right_on = '우측 열', how = '조인 방식')
# 후자의 방법은 병합 시의 기준 열의 이름이 다른 경우에 적용한다.
# inner는 교집합, outer는 합집합을 의미한다.
```

```python
Ans4
```

```python
Ans5 = pd.merge(df5, df6, on='Algeria', how='outer')
```
