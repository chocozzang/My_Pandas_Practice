```python
import pandas as pd
```

```python
dataURL = 'https://raw.githubusercontent.com/Datamanim/pandas/main/under5MortalityRate.csv'

df = pd.read_csv(dataURL)
```

```python
df.head(4)
```

```python
def remov_trust(x):
  return float(x.split('[')[0])

df['First Tooltip'] = df['First Tooltip'].apply(remov_trust)
df = df.drop('Indicator', axis = 1)

Ans = df
```

```python
Ans.head(4)
```

```python
Ans2 = df[(df['Dim1'] == "Both sexes") & (df['Period'] >= 2015)]
```

```python
Ans2.head(3)
```

```python
Ans3 = Ans2.pivot(index = 'Location', columns = 'Period', values = 'First Tooltip')

# pivot으로 데이터 프레임 만들기
# df.pivot(index = 행, columns = 열, values = 값) 으로 데이터 프레임을 만든다
```

```python
Ans3.head(3)
```

```python
Ans4 = df.pivot_table(index = 'Dim1', columns = 'Period', values = 'First Tooltip', aggfunc = 'mean')

# pivot table로 데이터 프레임 만들기
# df.pivot_table(index = 행, columns = 열, values = 값, aggfunc = 함수) 으로 피벗 테이블을 만든다
```

```python
Ans4.iloc[:, :4]
```
