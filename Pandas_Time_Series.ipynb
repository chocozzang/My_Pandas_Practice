```python
import pandas as pd
import datetime
```

```python
DataURL = "https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv"

df = pd.read_csv(DataURL)
```

```python
df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])

Ans = df.Yr_Mo_Dy
```

```python
Ans.head(4)
```

```python
Ans2 = df.Yr_Mo_Dy.dt.year.unique()

# 유일값 찾기 .unique()

# datetime 타입의 각 파트는 .dt로 접근할 수 있다.

# .date           YYYY-MM-DD
# .year           연도 (4자리)
# .month          월(숫자)
# .month_name()   월(문자)
# .day            일(숫자)
# .time           HH:MM:SS(숫자)
# .hour           시(숫자)
# .minute         분(숫자)
# .second         초(숫자)
```

```python
Ans2
```

```python
Ans3 = df

def my_year(myear):
  if myear.year >= 2061:
    return datetime.date(myear.year - 100, myear.month, myear.day)
  else:
    return datetime.date(myear.year, myear.month, myear.day)

Ans3.Yr_Mo_Dy = Ans3.Yr_Mo_Dy.apply(my_year)

# datetime은 모듈로 import하여 사용할 수 있다.

# datetime의 함수 중
# datetime.date는 2023-02-05와 같은 형식으로 저장하고
# datetime.datetime은 2023-02-05 18:16:40과 같은 형식으로 저장할 수 있는 함수이다.
```

```python
Ans3.head(4)
```

```python
Ans4 = df

Ans4 = Ans4.groupby(Ans4.Yr_Mo_Dy.dt.year).mean()

# datetime 타입인 경우에, .dt.year등으로 접근하여 연도/월/일 별 평균 등을 계산할 수 있다.
```

```python
Ans4.head(4)
```

```python
Ans5 = df

Ans5['weekday'] = Ans5.Yr_Mo_Dy.dt.weekday

Ans5 = Ans5['weekday'].to_frame()

# 요일을 알아보는 방법은 datetime.dt.weekday와 datetime.dt.day_name()이 있다.
# weekday는 요일을 숫자로 나타내는 것이고,
# day_name()은 요일을 문자로 나타내는 것이다.
```

```python
Ans5.head(3)
```

```python
def weekchk(x):
  if x <= 4: return 0
  else:      return 1

Ans6 = df

Ans6['weekday'] = Ans6.Yr_Mo_Dy.dt.weekday

Ans6['WeekCheck'] = Ans6.weekday.apply(weekchk)

Ans6 = Ans6['WeekCheck'].to_frame().head(3)
```

```python
Ans7 = df.groupby(df.Yr_Mo_Dy.dt.month).mean()
```

```python
Ans7.head(4)
```

```python
Ans8 = df

Ans8 = Ans8.fillna(method = "ffill")
Ans8 = Ans8.fillna(method = "bfill")

# 결측값을 채우는 방법 fillna() 함수를 활용한다.
# 이때, 값을 지정해서 채울 수도 있지만,
# method를 활용하면, 이전 값 혹은 이후 값으로 결측값을 채울 수도 있다.

# ffill/pad : 이전 값으로 채움
# bfill/backfill : 이후 값으로 채움

df = Ans8
```

```python
Ans8.isnull().sum()
```

```python
Ans9 = df.groupby(df.Yr_Mo_Dy.dt.to_period("M")).mean()

# dt.to_period : 원하는 부분까지만 추출할 수 있음
# ex) "Y" : 연도까지 "M" : 월까지 "D" : 일까지
```
