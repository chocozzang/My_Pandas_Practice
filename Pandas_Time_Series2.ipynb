```python
import pandas as pd
```

```python
DataURL = "https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv"

df = pd.read_csv(DataURL)
```

```python
df['(년-월-일:시)']
```

```python
def change_date(date):
  
  import datetime as dtt

  hour = date.split(':')[1]
  date = date.split(':')[0]
  
  if hour == "24": 
    hour = "00"
    new_dt = date + " " + hour + ":00:00"
    cng_dt = pd.to_datetime(new_dt) + dtt.timedelta(days = 1)

  else:
    new_dt = date + " " + hour + ":00:00"
    cng_dt = pd.to_datetime(new_dt)

  return cng_dt

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)

# 문자열을 datetime으로 변경하고자 한다면, apply와 함수를 활용할 수 있다.
```

```python
Ans7 = df.resample(rule = '1W', on = '(년-월-일:시)').agg(['min','max','mean','std'])

# 데이터를 주간 집계 하고자 한다면, resample을 활용할 수 있다.
# 이때, resample의 인자는 rule과 on이 있는데,
# rule은 간격을 설정하는 인자이고 on은 어떤 데이터에 대해서 resample할 것인지를 나타내는 인자이다.
```
